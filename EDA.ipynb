```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import pandas as pd
```

```python
#Descargamos el indice a analizar
sp500 = yf.download('^GSPC',start='1990-01-01')
sp500.reset_index(inplace=True)
```

```python
#Resumen de estad√≠sticas descriptivas
sp500.describe()
```

```python
sp500.columns
```

```python
# Heatmap de correlaciones entre variables num√©ricas
sns.heatmap(sp500.corr(), cmap='coolwarm', annot=True)
plt.show()
#De esta manera podemos ver en que grado se afectan las distintas variables
```

```python
# Histogramas de los precios de cierre
plt.hist(sp500['Close'],edgecolor = 'black')
plt.show()
```

```python
#grafico de linea del precio de cierre respecto al tiempo
sp500['Date'] = pd.to_datetime(sp500['Date'])
sp500.set_index('Date', inplace=True)
sp500['Close'].plot(figsize=(12, 6))
plt.show()
#Es un grafico ideal para ver la tendencia que tiene el indicie y como fue evolucionando en el tiempo
#asi como ver sus coatizaciones mas bajas, donde podemos ver que las mismas coinciden con las
#crisis economicas
```

```python
#Diagrama de caja de los precios de cierre
plt.boxplot(sp500['Close'],vert=False)

#Vemos aca que el diagrama de caja toma los valores mas altos como outlaiers, 
#esto se debe a que la cotizacion del sp500 a crecido exponencialmente respecto a como se venia 
#manejadndo el precio antiguamente.
#viendo el histograma, el diagrama de caja y el grafico de lineas podemos ver como la economia a 
#cambiado exponencialmente en las ultimas decadas
```

```python
#Gr√°fico de dispersi√≥n de la variable Volume respecto al tiempo
sns.set(style="darkgrid")
sns.scatterplot(x="Date", y="Volume", data=sp500)

# Mostrar el gr√°fico
plt.show()
```
